### **1 - CARREGANDO O ARQUIVO**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('tabelasBrasileirao.csv')
df
```

```python
df['%_GC'] = df['GC'] / df['GP']
```

```python
fd_dict = {
            'MG':['Cruzeiro', 'Atlético Mineiro','América Mineiro','Ipatinga'],
            'SP':['Santos', 'São Paulo', 'São Caetano','Guarani', 'Palmeiras', 'Corinthians', 'Ponte Preta', 'Portuguesa','Grêmio Barueri', 'Santo André','Grêmio Prudente','Red Bull Bragantino'],
            'RS':['Internacional', 'Grêmio','Juventude'], 
            'SC':['Joinville','Figueirense','Criciúma','Avaí','Chapecoense'],
            'RJ':['Flamengo', 'Fluminense', 'Vasco da Gama', 'Botafogo'],
            'GO':['Goiás', 'Atlético Goianiense'],
            'PR':['Paraná','Athletico Paranaense', 'Coritiba'],
            'CE':['Fortaleza','Ceará'],
            'PA':['Paysandu'],
            'BA':['Bahia', 'Vitória'],
            'DF':['Brasiliense'], 
            'PE': ['Santa Cruz', 'Sport', 'Náutico'],
            'RN':['América de Natal'], 
            'MT':['Cuiabá'],
            'AL':['CSA']
        }



for fd in fd_dict:
    df.loc[df['Equipes'].isin(fd_dict[fd]),'fd'] = fd
df
```

### **1. Quem foram os maiores campeões**

Abaixo temos um ranking do campeão com maior pontuação até com a menor
Obs.: Estamos usando todas as edições desde 2003. Lembrando que houve mudança na quantidade de equipes com o passar do tempo, vamos ir verificando isso com o passar da análise.

### 1.1 - Ranking geral

```python
df_campeoes.sort_values(by='P', ascending=False)
```

#### 1.1 - Observações

Notamos que, quando vemos todas as edições (mesmo com número de jogos diferentes), temos o **Cruzeiro** com a maior pontuação, seguido do **Flamengo**, sendo este título no ano de _2019_ com _38_ jogos totais, um feito muito bom sendo que o **Cruzeiro** teve _46_ jogos; quando olhamos o **terceiro** colocado, temos o **Santos**, que foi campeão com o mesmo número de jogos que o **Cruzeiro**, o **Corinthians** que foi campeão em _2005_, tendo _42_ jogos, esta em **_6º_**, tendo a mesma pontuação de sua campanha de _2015_ com a mesma pontuação.

As **%** (porcentagens) não possuem relação com as _potuações_ e _quantidade de jogos_. **Flamengo** de 2019 tem a melhor **%** de todos os clubes, inclusive que o **Santos** de _2004_ e **Corinthians** de _2005_.

Gols Marcados e sofridos (GP e GC
Então, podemos dizer que:
- Número de jogos totais não implica em boas campanhas.
- Porcentagens não tem relação com pontuação e quantidade de jogos.

### 1.2 - Ranking por número de jogos

```python
columns = list(df_campeoes.columns)
columns.remove("Pos.")
columns.remove("Classificação ou rebaixamento")
columns
```

```python
for numero_jogos in df_campeoes['J'].unique():
    print(f'Quantidade de jogos no campeonato: {numero_jogos}')
    print(df_campeoes[columns].groupby(by='J').get_group(numero_jogos).sort_values(by='P', ascending=False))
    print('\n')
```

#### 1.2 - Observações

Com a separação por número de jogos, notamos um padrão. Se observarmos os piores campeões, notamos que eles possuem metade do número de jogos como _vitórias_ (aproximadamente);

Aparenta ter um relação entre **GC** e **D** (Derrotas), quanto maior a **%_GC** maior o número de **derrotas**

- Ter metade ou mais de _vitórias_ com relação ao número de jogos é o requisito mínimo para ser campeão.
- Relação entre **%_GC** e **número de derrotas**

```python
df_campeoes.groupby(by='fd').count().sort_values(by='Pos.', ascending=False)
```

```python
for numero_jogos in df_campeoes['J'].unique():
    print(f'Quantidade de jogos no campeonato: {numero_jogos}')
    df_group = df_campeoes[['Equipes', 'J']].groupby(by='J').get_group(numero_jogos)['Equipes'] 
    print(f'Quantidade de campeoes totais : {df_group.count().sum()}')
    print('Tabela de quantidade de títulos por Equipe')
    print(df_campeoes[columns].groupby(by='J').get_group(numero_jogos).groupby(by='Equipes').count())
    print('Tabela de quantidade de títulos por FD')
    print(df_campeoes[columns].groupby(by='J').get_group(numero_jogos).groupby(by='fd').count())
    print('\n')
```

#### 1.3 - Observação

Fora realizados _19_ edições do Campeonato Brasileiro no formato de pontos corridos (2003 - 2021), delas, somente _8_ equipes sagraram-se campeãs.

No **ranking geral**, notamos que o **Corinthians** é o maior acumulador de títulos neste formato, com _4_.
Quando vamos a separação por número de jogos, notamos que o **Santos** é a única equipe que não aparece mais após a fixação em _38_ jogos.

Também notamos uma certa hegemonia regional entre **SP**, **RJ** e **MG**, sendo **SP** a região com maior número de equipes campeãs e maior quantidade de títulos.
